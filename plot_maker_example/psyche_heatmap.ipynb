```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
# Helvetica 
rcParams['font.family'] = 'Helvetica'

# CSV  
df = pd.read_csv(
    '/Users/inw/Desktop//matplotlib/data_psyche/plot _7.csv')

# 'Appropriateness'   
df.set_index('Conformity', inplace=True)

#   (transpose)
df = df.T

#   
plt.figure(figsize=(16, 11))

#   (annot fmt  )
ax = sns.heatmap(df, cmap='Blues', vmin=0, vmax=100, square=True,
                 linewidths=0.5, cbar=False, annot=True, fmt='.0f', annot_kws={"fontsize": 10})

# y   
ax.yaxis.tick_right()
ax.yaxis.set_label_position('right')

# x  ,    
plt.xticks(rotation=90, ha='center', fontsize=16)

# y     
plt.yticks(rotation=0, fontsize=16)

plt.title('Conformity Heatmap by Elements', fontsize=24, pad=20)


#   
# [left, bottom, width, height]
cbar_ax = plt.gcf().add_axes([0.7, 0.08, 0.4, 0.02])
cbar = plt.colorbar(ax.collections[0], cax=cbar_ax, orientation="horizontal")

#   
cbar.ax.tick_params(labelsize=16)  # Tick   
cbar.outline.set_visible(False)   #   
cbar.set_label('Conformity (%)', fontsize=16)  #   


# #  
plt.tight_layout()
# #  
# plt.subplots_adjust(top=0.9, bottom=0.2, left=0.1, right=0.9)


#   
plt.savefig('/Users/inw/Desktop//matplotlib/figure_psyche/appropriateness_heatmap_revised_label_percent_category_colorbar_colorblind.png',
            dpi=300, bbox_inches='tight')

#  
plt.show()
```
